```python
import pandas as pd
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns

#--- suppress warnings ---
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read CSV
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded["Tasmania - Metro Hobart _GTFS v2.csv"]),sep=',')
df.head()
```

```python
df.isnull().sum()
```

```python
def clean_column_name(column_names):
    clean_column_names = []
    for c in column_names:
        c = c.lower().replace(' ','_')
        clean_column_names.append(c)
    return clean_column_names    

df.columns = clean_column_name(df.columns)
```

```python
df.columns = ['stop_id', 'stop_name', 'latitude', 'longtitude', 'mode', 'state_code'] 
df.columns
```

```python
df.info()
```

```python
duplicate = df[df.duplicated()]
print("Duplicate Rows :")
duplicate
```

```python
#--- take the means to replace the zero in baths & bed ---
df['1_br_flat'] = df['1_br_flat'].replace([np.nan], df['1_br_flat'].median())
df['2_br_flat'] = df['2_br_flat'].replace([np.nan], df['2_br_flat'].median())
df['3_br_flat'] = df['3_br_flat'].replace([np.nan], df['3_br_flat'].median())
df['4_br_flat'] = df['4_br_flat'].replace([np.nan], df['4_br_flat'].median())
df['1_br_house'] = df['1_br_house'].replace([np.nan], df['1_br_house'].median())
df['2_br_house'] = df['2_br_house'].replace([np.nan], df['2_br_house'].median())
df['3_br_house'] = df['3_br_house'].replace([np.nan], df['3_br_house'].median())
df['4_br_house'] = df['4_br_house'].replace([np.nan], df['4_br_house'].median())
df['state'] = df['state'].replace([np.nan], ['Brisbane'])
```

```python
df.isnull().sum()
```

```python
# Replace NaN with zero on all columns 
df = df.fillna(0)
```
